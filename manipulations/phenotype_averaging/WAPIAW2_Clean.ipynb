```python
# Finding unique missing values - up to 5, also including none missing and assigning to value
x_1 = TNS_NotMissing['NotMissing'].unique()
x_1.sort()
x_1
```

```python
# Finding counts for the unique missing values 
y_1 = (TNS_NotMissing['NotMissing']).value_counts().sort_index()
y_1
```

```python
# Plotting missing value counts 
import matplotlib.pyplot as plt
plt.bar(x_1, y_1, color ='maroon',width = 0.7) 
plt.title('Distribution of Present Data for Neuroticism')
plt.legend('Count')
plt.xlabel('Number of present data across 4 Instances')
plt.ylabel('Present Data Count')
plt.savefig('plot0.png', dpi=300, bbox_inches='tight')
plt.show() 
```

```python
# Counting number of present sums across the is 4 instances 
TNS_Missing = Overall_Neuroticism_Table.isna().T.sum()
TNS_Missing = pd.DataFrame(TNS_Missing, columns = ['MissingData'])
TNS_Missing
```

```python
# Review columns in the table 
phenotype_tableNew.columns
```

```python
# From the above select columns of interest to append to the main dataframe 
A = Instance_0['Sum_0'] # 20127-0.0
B = Instance_1['Sum_1'] # 20127-1.0
C = Instance_2['Sum_2'] # 20127-2.0 
D = Instance_3['Sum_3'] # 20127-3.0
E = phenotype_table_IQ['Average_IQ'] # 20016_Mean
F = Overall_Neuroticism_Table['Average_NeuroticismScore'] # 20017_Mean
G = TNS_NotMissing # 20016-Avail 
H = TIQ_NotMissing # 20017-Avail
```

```python
# Renaming columns
TableToAppend.columns = ['20127-0.0','20127-1.0','20127-2.0','20127-3.0','20016_Mean','20127_Mean','20016-Avail ','20127-Avail']
TableToAppend 
```
